```python
import os
from glob import glob
from xml.etree import ElementTree as ET


path_fish_remix2=sorted(glob("./dataset/test_data(label)/fish_remix2/xml/*.xml"))

def get_data(path):
    for i in range(len(path)):
        filename=os.path.basename(path[i].split('.')[1])
        with open(path[i],"r") as file:
            tmp=[]
            tree=ET.parse(file)
            root = tree.getroot()
            for obj in root.iter('object'):
                name = obj.find('name').text
                if name=="fish1":
                    name="clownfish1"
                if name=="fish2":
                    name="clownfish2"
                if name=="fish3":
                    name="clownfish3"
                xmlbox = obj.find('bndbox')
                b = (int(xmlbox.find('xmin').text), int(xmlbox.find('ymin').text), int(xmlbox.find('xmax').text), int(xmlbox.find('ymax').text))
                tmp.append(str(name)+" "+" ".join(str(a) for a in b)+"\n")
        with open('./mAP/input/ground-truth/'+filename+'.txt',"w") as file2:
            file2.writelines(tmp)

get_data(path_fish_remix2)
```
