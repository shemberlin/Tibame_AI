```
import glob
from xml.etree import ElementTree as ET
list_file = open('anno.txt', 'w') 
fish1=glob.glob("drive/MyDrive/Colab Notebooks/Yolo_v3/dataset/train_image/fish1/xml/*.xml")
fish2=glob.glob("drive/MyDrive/Colab Notebooks/Yolo_v3/dataset/train_image/fish2/xml/*.xml")
fish3=glob.glob("drive/MyDrive/Colab Notebooks/Yolo_v3/dataset/train_image/fish3/xml/*.xml")
fish_remix1=glob.glob("drive/MyDrive/Colab Notebooks/Yolo_v3/dataset/train_image/fish_remix1/xml/*.xml")
# fish_remix2=glob.glob("drive/My Drive/Colab Notebooks/Yolo_v3_tmp/fish_remix2/xml/*.xml")
# ()
def get_data(fname):
  total=len(fname)
  print(total)
  for i in range(total):
    with open(fname[i],"r") as f:
      print(f)
      file_id = ''.join(fname[i].split('/')[-1].split('.')[:-1])
      fish_name = ''.join(fname[i].split('/')[6])
      list_file.write(f'/dataset/train_image/{fish_name}/img/{file_id}.png')
      et=ET.parse(f)
      root = et.getroot()
      for obj in root.iter('object'):
        name = obj.find('name').text
        xmlbox = obj.find('bndbox')
        b = (int(xmlbox.find('xmin').text), int(xmlbox.find('ymin').text), int(xmlbox.find('xmax').text), int(xmlbox.find('ymax').text))
        # print(name)
        # print(b)
        if name=="fish3":
          name=2
        elif name=="fish1":
          name=0
        else:
          name=1
        
        list_file.write(" " + ",".join([str(a) for a in b])+ "," +str(name))
        
      list_file.write('\n')    
get_data(fish1)
get_data(fish2)
get_data(fish3)
get_data(fish_remix1)
# get_data(fish_remix2)
```
